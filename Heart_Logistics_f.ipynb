```python
data_1 = raw_data[raw_data['RestingBP']>90]
data_1 = raw_data[raw_data['RestingBP']<=200]
data_1 = raw_data[raw_data['Cholesterol']>100]
data_1 = raw_data[raw_data['Cholesterol']<=420]
data_1.describe(include = 'all')
```

```python
#create dummy variables
data_2 = pd.get_dummies(data_1,drop_first = True)
data_2 = data_2.reset_index(drop = True)
data_2.head()
```

```python
targets = data_2['HeartDisease']
inputs = data_2.drop(['HeartDisease'],axis = 1)
inputs.head()
```

```python
#standardize the data
from sklearn import preprocessing 
scaled_inputs = preprocessing.scale(inputs)
scaled_inputs
```

```python
#shuffle the data
shuffled_indices = np.arange(scaled_inputs.shape[0])
np.random.shuffle(shuffled_indices)
shuffled_inputs = scaled_inputs[shuffled_indices]
shuffled_targets = targets[shuffled_indices]
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(scaled_inputs,targets,test_size = 0.2,random_state = 365)
```

```python
#Logistic Regression
import statsmodels.api as sm
reg_log = sm.Logit(y_train,x_train)
results_log = reg_log.fit()
```

```python
results_log.summary()
```

```python
cm = np.array(cm_df)
accuracy = (cm[0,0]+cm[1,1])/cm.sum()
accuracy
```

```python
#split the dataset into train, validation, and test
sample_count = shuffled_inputs.shape[0]
train_count = int(0.8*sample_count)
validation_count = int(0.1*sample_count)
test_count = int(sample_count - validation_count - train_count)
print(sample_count,train_count,validation_count,test_count)
```

```python
train_inputs = shuffled_inputs[:train_count]
validation_inputs = shuffled_inputs[train_count:validation_count + train_count]
test_inputs = shuffled_inputs[validation_count + train_count :]

train_targets = shuffled_targets[:train_count]
validation_targets = shuffled_targets[train_count:validation_count + train_count]
test_targets = shuffled_targets[validation_count + train_count :]
```

```python
#save the data in .npz files
np.savez('Heart_data_train',inputs = train_inputs,targets = train_targets)
np.savez('Heart_data_validation',inputs = validation_inputs,targets = validation_targets)
np.savez('Heart_data_test',inputs = test_inputs,targets = test_targets)
```
